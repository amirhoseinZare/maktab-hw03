```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

DATASET_PATH = './diabetes.csv'
```

```python

dataset = pd.read_csv(DATASET_PATH)

print(dataset.isna().sum())

dataset.fillna(dataset.mean(), inplace=True)

print(dataset.isna().sum())

```

```python
print(dataset.head())
```

```python
selected_columns = ['Glucose', 'BloodPressure', 'BMI', 'Age', 'Insulin', 'Outcome']
selected_data = dataset[selected_columns]

corr = selected_data.corr()

print(corr)
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")

plt.title("Heatmap - Correlation between main columns")
plt.show()
```

```python
outcome_corr = corr['Outcome']
outcome_corr = outcome_corr.drop('Outcome')
max_corr_feature = outcome_corr.idxmax()
max_corr_value = outcome_corr.max()

print(max_corr_feature, max_corr_value)
```
